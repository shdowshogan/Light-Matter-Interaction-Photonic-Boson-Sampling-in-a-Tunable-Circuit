"Now I'll discuss the experimental results of the paper, which is really where the central claim gets tested.

Up to this point in the presentation, we have already seen two sides of the story. On one side, Alice has the theoretical machinery. She characterizes the interferometer, obtains the measured unitary matrix, and from that she computes the expected Boson Sampling visibilities. On the other side, Bob actually runs the optical experiment using photons and measures what comes out of the circuit.

So the natural question now is: how do the authors decide whether Bob is actually doing what Boson Sampling predicts?

The answer is that they compare Alice's predicted visibilities with Bob's measured visibilities using an average L1-norm distance over all tested output configurations.

The formula they use is the average over all outputs T of the absolute difference between Alice's visibility and Bob's visibility. So if Alice and Bob agree very well, this L1 distance will be small. If they disagree strongly, the distance will be large.

I want to make one point clear here, because it matters for how we present our notebook. These final L1 comparison values are given directly in the paper. But the PDF does not provide a machine-readable table of Bob's raw visibility values for every output configuration. In the figures, Bob's results are shown as bars, but the exact numerical data behind all those bars are not tabulated in the paper.

So for this section, what we are doing is not claiming that we reconstructed Bob's entire raw dataset from scratch. Instead, we are faithfully discussing the comparison values that the authors themselves report in the paper, and interpreting what those values mean physically."

"Let me start with the two-photon case.

As a baseline test, the authors inject two photons into the circuit and look at all possible non-colliding output combinations. Since there are six output modes, the total number of non-colliding two-photon outputs is six choose two, which is fifteen.

For each of those outputs, Alice computes the visibility from the measured unitary, and Bob measures the visibility experimentally. Then they compare the two sets.

The result is very strong. The average L1 distance is reported as 0.027.

That is a very small number, and it tells us that Alice's theoretical predictions and Bob's experimental measurements match extremely well in the two-photon regime.

But the authors do not stop there, because agreement alone is not enough. A skeptic might still ask whether Bob is just reproducing something that could also arise from a classical optical field.

To test that, the paper introduces what I would call the classical sanity check. Alice also computes what the visibilities would look like if she used coherent-state inputs instead of ideal Fock-state photons. These coherent-state predictions are the classical-wave baseline.

Now if Bob's measured results were also close to the coherent-state prediction, then the experiment would not really establish anything strongly nonclassical. But that is not what happens.

The paper reports that the distance between Bob's measured visibilities and Alice's coherent-state prediction is 0.548.

That is much larger than 0.027. So the message is very clear: Bob is not just reproducing a generic classical interference pattern of the same interferometer. His results match the bosonic prediction much better than the coherent-state one.

And the authors explicitly say that this large disagreement shows the quantum distribution is sufficiently different from the classical one, meaning the chosen unitary is not some special easy case that a classical description would accidentally mimic.

So already in the two-photon case, the experiment is doing exactly what it is supposed to do. It validates the permanent-based prediction and separates it from the classical coherent-state alternative."

"Now let us move to the three-photon case, which is much more interesting and also much more demanding experimentally.

When the authors scale up from two photons to three photons, the system becomes more complex in two different senses. First, there are more interfering multi-photon paths contributing to each output. Second, the experimental difficulty increases significantly because the source and detection imperfections become more important.

Again, the authors compare Alice's predicted visibilities with Bob's measured visibilities across the non-colliding outputs.

This time, the reported L1 distance is 0.122.

So compared to the two-photon case, the agreement becomes worse, but it is still clearly good enough to support the Boson Sampling model.

The authors also explain why this distance is larger. They attribute it mainly to the increased ratio of higher-order photon emissions in the three-photon experiment. This makes sense physically, because the downconversion source becomes more problematic as the photon number increases. It is harder to approximate the ideal input state, and the errors start to show up more clearly.

But once again, the important comparison is not only Alice versus Bob. The coherent-state benchmark is also included.

For the three-photon case, the distance between Bob's measured results and Alice's coherent-state predictions is reported as 0.358.

So even though the three-photon experiment is less accurate than the two-photon one, it is still much closer to the bosonic prediction than to the classical coherent-state baseline. That means the genuinely quantum interference is still clearly present.

This is a very important part of the paper, because it shows that the Boson Sampling signature survives even when the experiment becomes more difficult and more realistic imperfections are present."

"After that, the paper goes one step further and studies colliding configurations.

Until this point, most of the analysis is focused on non-colliding events, meaning outputs where at most one photon exits each output mode. But Boson Sampling in principle includes colliding outputs as well, where two photons can emerge in the same output mode.

The authors therefore test those cases too. Experimentally, this requires photon-number resolution, because the setup must be able to distinguish between one photon in a mode and two photons in the same mode. In the paper, they achieve this using a 50-50 fibre beam splitter and an extra detector.

Now this is a more demanding test, because the output configurations are more complicated. But the result is again very encouraging.

For these colliding-output states, the paper reports an L1 distance of 0.153 between Alice's predicted visibilities and Bob's measured visibilities.

So even here, the agreement is still quite good.

And once again, the classical sanity check makes the picture even stronger. The distance between Bob's measurements and Alice's coherent-state predictions becomes extremely large, namely 0.995.

That is almost maximal compared to the earlier cases. So in the colliding-output regime, Bob's results are still close to the bosonic prediction, but very far from the classical coherent-state baseline.

This is important because it shows that the robustness of the Boson Sampling description is not limited to the simplest output class. It continues to hold even in these more complex colliding configurations."

"So overall, across all three tests, the pattern is very consistent.

For two photons, Alice and Bob agree extremely well, with L1 equal to 0.027.

For three photons, the agreement is still good, with L1 equal to 0.122.

For colliding outputs, the agreement remains strong, with L1 equal to 0.153.

And in every one of these cases, Bob's results are much farther from the coherent-state prediction than from the bosonic prediction.

That is really the central experimental conclusion of this part of the paper."

"Now let me move to the final part of this section, which is about imperfections and the path forward.

A careful reader will immediately notice that the agreement is not perfect. The L1 distance is not zero, and it gets worse as we go from two photons to three photons.

So the obvious question is: if Alice's theory is correct, why doesn't Bob match it exactly?

The answer is that Alice is calculating in an ideal mathematical world, while Bob is working with real hardware.

Alice assumes perfectly indistinguishable Fock-state inputs. Bob, however, generates photons using spontaneous parametric downconversion, or SPDC. And SPDC is probabilistic, not ideal.

This introduces two major sources of noise.

The first is higher-order emission. Even if the source is tuned to produce single pairs, there is always some probability that it produces extra photons. Those extra photons contaminate the intended input state.

The second is distinguishability. The photons produced by downconversion can be spectrally entangled, which means they are not perfectly identical. As soon as the photons become even slightly distinguishable, the pure bosonic interference is weakened.

This explains why the experimental agreement becomes worse in the three-photon case. Bob is trying to realize an ideal input state, but the source itself is imperfect."

"This leads to a very important hardware trade-off.

If Bob increases the pump power, then he can generate photons more often, which improves the count rate and makes data collection faster.

But increasing pump power also increases higher-order emissions. So the source becomes noisier, and the output statistics are pushed away from the ideal quantum distribution and toward a more classical-looking regime.

That is exactly what Figure 5 in the paper shows.

So the experiment has a built-in conflict. Low pump power gives cleaner photons but very low rates. High pump power gives better rates but introduces more noise.

And because SPDC is fundamentally probabilistic, the problem becomes much worse as the number of photons grows. The success probability drops rapidly with photon number."

"This brings us to the final big-picture point.

This paper does not yet disprove the Extended Church-Turing thesis by itself. The system is still too small. Two and three photons are not enough to enter the truly classically intractable regime.

But what the paper does achieve is very important. It validates the foundational physical mechanism of Boson Sampling.

It shows that a real photonic device, using a measured interferometer unitary, produces output statistics that follow the permanent-based bosonic prediction and not the classical coherent-state baseline.

That is the experimental foundation the field needs before trying to scale up.

According to the theoretical estimates discussed in the paper, the really decisive regime would be around 20 to 30 photons in a network with many more modes, where Bob could sample quickly but Alice could no longer verify the result efficiently on a classical computer.

Reaching that regime is no longer mainly a question of theory. It becomes a question of engineering: better photon sources, lower losses, and more efficient detectors.

So if I summarize this entire section in one sentence, it would be this:

the experiment already confirms the correctness of the Boson Sampling mechanism in small systems, but the final challenge is to scale it into the regime where classical verification breaks down and the Extended Church-Turing thesis can be tested directly."

"And that is why these experimental results matter so much. They do not end the race yet, but they show that Bob is clearly running in the right direction."
